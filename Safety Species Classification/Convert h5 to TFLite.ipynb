```python
import tensorflow as tf
```

```python
# Load model .h5
model_species = tf.keras.models.load_model('equipments.h5')

# Convert to TensorFlow Lite
converter = tf.lite.TFLiteConverter.from_keras_model(model_species)
tflite_model = converter.convert()

with open('species_model.tflite', 'wb') as f:
    f.write(tflite_model)

print("Succes")
```
